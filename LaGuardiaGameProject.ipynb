```python
import random

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(player1, player2):
    if player1 == player2:
        return "tie"
    elif (player1 == "rock" and player2 == "scissors") or \
         (player1 == "scissors" and player2 == "paper") or \
         (player1 == "paper" and player2 == "rock"):
        return "Cherry"
    else:
        return "Larry"

def play_game():
    cherry_score = 0
    larry_score = 0

    for _ in range(5):
        cherry_choice = input("Cherry, enter your choice (rock, paper, scissors): ").lower()
        larry_choice = get_computer_choice()
        print(f"Larry chose: {larry_choice}")

        winner = determine_winner(cherry_choice, larry_choice)
        if winner == "Cherry":
            cherry_score += 1
        elif winner == "Larry":
            larry_score += 1

        round_winner = winner if winner != "tie" else "It's a tie!"
        print(f"Round winner: {round_winner}")

    print("Final Scores:")
    print(f"Cherry: {cherry_score}")
    print(f"Larry: {larry_score}")

    if cherry_score > larry_score:
        print("Congratulations Cherry! You win the game!")
    elif larry_score > cherry_score:
        print("Congratulations Larry! You win the game!")
    else:
        print("The game is a tie!")

    play_again = input("Do you want to play again? (yes/no): ").lower()
    if play_again == "yes":
        play_game()
    else:
        print("Thanks for playing! Goodbye!")

def main():
    print("Welcome to Rock, Paper, Scissors!")
    play_game()

if __name__ == "__main__":
    main()
```
